## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# let's inspect client with id '66037'
client_df[client_df['client_id'] == 66037]
```

```python
# Create a column that calculates the subtotal for each line using the unit_price and the qty
client_df['line_subtotal'] = client_df['unit_price'] * client_df['qty']
client_df.head()
```

```python
# Create a column that calculates the total_weight by multiplying the qty and unit_weight
client_df['total_weight'] = client_df['qty'] * client_df['unit_weight']
client_df.head()
```

```python
# Create a column for shipping price.
# Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
client_df['shipping_price'] = client_df['total_weight'].apply(lambda x: x * 7 if x > 50 else x * 10)
client_df
```

```python
# Create a column for the total price using the subtotal and the shipping price along with a sales tax of 9.25% and round to 2 decimal places
client_df['line_price'] = (client_df['line_subtotal'] + client_df['shipping_price']) * 1.0925

# Round the totals_price column to 2 decimal places
client_df['line_price'] = client_df['line_price'].round(2)
client_df.head()
```

```python
# Create a column for the cost of each line using unit cost, qty, and
# shipping price (assume the shipping cost is exactly what is charged to the client).
client_df['line_cost'] = client_df['unit_cost'] * client_df['qty'] + client_df['shipping_price']
client_df.head()

```

```python
# Create a column for the line_profit of each line using line_cost and line_price
client_df['line_profit'] =  client_df['line_price'] - client_df['line_cost']
client_df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity spend? Check your work from Part 1 for client ids. The top 5 clients
# are stored in the variable top_5_clients

top_5_clients
```

```python
# Here we're looping through the each client_id of the list top_5_clients and calculating the sum of the line_price
for client_id in top_5_clients:
    print(f"{client_id}: ${client_df.loc[client_df['client_id'] == client_id, 'line_price'].sum():.2f}")
```

```python
# Create a summary DataFrame showing the totals for the for the top 5 clients with the following information:
# total units purchased, total shipping price, total revenue, and total profit. 
top_5_clients_df = pd.DataFrame({
    'client_id': top_5_clients,
    'qty': [client_df.loc[client_df['client_id'] == clientID, 'qty'].sum() for clientID in top_5_clients],
    'shipping_price': [client_df.loc[client_df['client_id'] == clientID, 'shipping_price'].sum() for clientID in top_5_clients],
    'line_price': [client_df.loc[client_df['client_id'] == clientID, 'line_price'].sum() for clientID in top_5_clients],
    'line_cost': [client_df.loc[client_df['client_id'] == clientID, 'line_cost'].sum() for clientID in top_5_clients],
    'line_profit': [client_df.loc[client_df['client_id'] == clientID, 'line_profit'].sum() for clientID in top_5_clients]
})  
top_5_clients_df

```

```python
# Format the data and rename the columns to names suitable for presentation.
# Define the money columns. 
top_5_clients_reformatted = top_5_clients_df.rename(columns={
    'client_id': 'Client ID',
    'qty': 'Units',
    'shipping_price': 'Shipping Price (millions)',
    'line_price': 'Total Revenue (millions)', 
    'line_cost': 'Total Cost (millions)', 
    'line_profit': 'Total Profit (millions)'
    })

# Define a function that converts a dollar amount to millions.
def currency_format_millions(x):
    return f"{x/1000000:.6f}"


# Apply the currency_format_millions function to only the money columns
top_5_clients_reformatted['Shipping Price (millions)'] = top_5_clients_reformatted['Shipping Price (millions)'].apply(currency_format_millions)
top_5_clients_reformatted['Total Revenue (millions)'] = top_5_clients_reformatted['Total Revenue (millions)'].apply(currency_format_millions)
top_5_clients_reformatted['Total Cost (millions)'] = top_5_clients_reformatted['Total Cost (millions)'].apply(currency_format_millions)
top_5_clients_reformatted['Total Profit (millions)'] = top_5_clients_reformatted['Total Profit (millions)'].apply(currency_format_millions)

# Rename the columns to reflect the change in the money format. 
top_5_clients_reformatted

```

```python
# Sort the updated data by "Total Profit (millions)" form highest to lowest and assign the sort to a new DatFrame.
top_5_clients_profit_df = top_5_clients_reformatted.sort_values(by='Total Profit (millions)', ascending=False)
top_5_clients_profit_df
```
